```python
#Задача 1: Подсчитать произведение ненулевых элементов на диагонали прямоугольной матрицы.
#Например, для X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]) ответ – 3.
#

def prod_non_zero_diag(x):
    product = 1
    for i in range(min(x.shape)):
        if x[i, i] != 0:
            product *= x[i, i]
    return product

def prod_non_zero_diag_vectorized(x):
    diag = x.diagonal()
    return diag[diag != 0].prod()

X = np.array([[1,0,1],[2,0,2],[3,0,3],[4,4,4]])
print("Невекторизованная реализация: ",prod_non_zero_diag(X))
print("Векторизованная реализация: ", prod_non_zero_diag_vectorized(X))
    
```

```python
#Задача 2: Даны два вектора x и y. Проверить, задают ли они одно и то же мультимножество.
#Например, для x = np.array([1, 2, 2, 4]), y = np.array([4, 2, 1, 2]) ответ – True.
#

def are_multisets_equal(x, y):
    x_array = x.tolist()
    y_array = Y.tolist()
    x_array.sort()
    y_array.sort()
    return x_array == y_array

def are_multisets_equal_vectorized(x, y):
    return np.array_equal(np.bincount(x), np.bincount(y))
X = np.array([1,2,2,4])
Y = np.array([4,2,1,2])
print("Невекторизованная реализация: ", are_multisets_equal(X, Y))
print("Векторизованная реализация: ", are_multisets_equal_vectorized(X, Y))
```

```python
#Задача 3: Найти максимальный элемент в векторе x среди элементов, перед которыми стоит нулевой.
#Например, для x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0]) ответ – 5.
#

def max_after_zero(x):
    max_after_zero = -1
    for i in range(len(x) -1):
        if (x[i] == 0) and (x[i+1] > max_after_zero):
            max_after_zero = x[i+1]
    return max_after_zero
        
def max_after_zero_vectorized(x):
    #x1 = x[:-1]
    #x2 = x[1:]
    #after_zero_array = np.where(x1 == 0, x2, np.NaN)
    #return np.nanmax(after_zero_array)
    m = x[(np.array(np.nonzero(x[:-1] == 0)) + 1)]
    if m.size == 0:
        return 0
    else:
        return m.max()

X = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0])

print("Невекторизованная реализация ", max_after_zero(X))
print("Векторизованная реализация ", max_after_zero_vectorized(X))
```

```python
#Задача 4: Дан трёхмерный массив, содержащий изображение, размера (height, width, numChannels),
#а также вектор длины numChannels.
#Сложить каналы изображения с указанными весами, и вернуть результат в виде матрицы размера (height, width).
#В ноутбуке приведите пример работы функции – преобразуйте цветное изображение в оттенки серого,
#использовав коэффициенты np.array([0.299, 0.587, 0.114]).
#Считать реальное изображение можно при помощи функции scipy.misc.imread
#(если изображение не в формате png, установите пакет pillow).
#Обратите внимание, что в изображении может быть не три канала!
#За решения, явно или неявно использующие трёхканальность изображения будет присуждена половина баллов.
#
def convert_image(img, coefs):
    height = img.shape[0]
    width = img.shape[1]
    numChannels = img.shape[2]
    assert numChannels == len(coefs)
    
    result = [[0  for _ in range (width) ] for _ in range(height)]
    
    for y in range(height):
        for x in range(width):
            for channel in range(numChannels):
                result[y][x] += img[y][x][channel] * coefs[channel]
    return result
def convert_image_vectorized(img, coefs):
    return img @ coefs
import matplotlib.pyplot as plt
from imageio import imread

img = imread("Simbiosis.png")
weight = np.array([0.299, 0.587, 0.114])

img_gray = convert_image(img, weight)
img_gray_vectorized = convert_image_vectorized(img, weight)
figure, axeses = plt.subplots(nrows=1, ncols=3, figsize=(20, 20))

axeses[0].imshow(img)
axeses[0].grid()

axeses[1].imshow(img_gray, cmap="gray")
axeses[1].grid()

axeses[2].imshow(img_gray_vectorized, cmap="gray")
axeses[2].grid()
plt.show()
```

```python
#Задача 5: Реализовать кодирование длин серий (Run-length encoding).
#Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины.
#Первый содержит числа, а второй - сколько раз их нужно повторить.
#Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) ответ – (np.array([2, 3, 5]), np.array([3, 3, 1])).
#

def run_length_encoding(x):
    
    elements = []
    counters = []
    
    current_element = None
    current_counter = None
    
    for element in x:
        if current_element == element:
            current_counter += 1
        else:
            if current_element != None:
                elements.append(current_element)
                counters.append(current_counter)
            current_element = element
            current_counter = 1
    if current_element != None:
        elements.append(current_element)
        counters.append(current_counter)
    return elements, counters

def run_length_encoding_vectorized(x):
    n = len(x)
    if n == 0:
        return (np.array([], dtype = int),
                np.array([], dtype = x.dtype))
    starts = np.r_[0, np.flatnonzero(x[1:] != x[:-1]) + 1]
    lengths = np.diff(np.r_[starts, n])
    values = x[starts]
    return values, lengths
    
X = np.array([2, 2, 2, 3, 3, 3, 5])

print("Невекторизованная реализация: ", run_length_encoding(X))
print("Векторизованная реализация: ", run_length_encoding_vectorized(X))
```

```python
%pylab inline

import seaborn as sns
sns.set(style="whitegrid")

from sklearn import datasets
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import pandas as pd
```

```python
def pairplot(df, target):
    ncol, nrow = 7, df.shape[1] // 7 + (df.shape[1] % 7 > 0)
    plt.figure(figsize=(ncol * 4, nrow * 4))

    for i, feature in enumerate(df.columns):
        plt.subplot(nrow, ncol, i + 1)
        plt.scatter(df[feature], target, s=10, marker='o', alpha=.6)
        plt.xlabel(feature)
        if i % ncol == 0:
            plt.ylabel('target')
```

В данном задании мы рассмотрим стандартный датасет для задачи регрессии Boston Housing, в котором необходимо предсказать стоимость недвижимости по 13 признакам.

Датасет достаточно просто загрузить из библиотеки scikit-learn:

```python
data = datasets.load_boston()
df = pd.DataFrame(data.data)
df.columns = data.feature_names
y = data.target
```

__Задание 1.__ Обучение линейной регрессии.

```python
model_linreg = LinearRegression().fit(x_train, y_train)
model_lasso = Lasso().fit(x_train, y_train)
model_ridge = Ridge().fit(x_train, y_train)

model_linreg_score = model_linreg.score(x_test, y_test)
model_lasso_score = model_lasso.score(x_test, y_test)
model_ridge_score = model_ridge.score(x_test, y_test)
print("Линейная регрессия:", model_linreg_score)
print("Линейная регрессия L1 регуляризацией:", model_lasso_score)
print("Линейная регрессия L2 регуляризацией:", model_ridge_score)
```
